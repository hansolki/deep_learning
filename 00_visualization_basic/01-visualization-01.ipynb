# subplot
- plot 여러 번 실행 후 plt.show() => 한 그래프 내부에 여러 직선
- subplot => 행/열 단위로 묶어서 여러 데이터 병행표시

````python
plt.plot(np.arange(5)**2)
plt.title('char1')
plt.plot(np.arange(10), np.arange(10)**3, 'r--')
plt.title('chart2')
plt.show()

#=> 한 그래프안에 선은 2개, 제목은 덮어씌워짐
````

````python
plt.subplot(2, 1, 1) #=> 행수, 열수, 해당순서
plt.plot(np.arange(5)**2)
plt.title('char1')
plt.subplot(2, 1, 2)
plt.plot(np.arange(5), np.arange(5)**3, 'bs--')
plt.title('chart2')
plt.tight_layout() #=> 두 차트 사이 간격 주기
plt.show()
````

````python
plt.subplot(1, 2, 1) #=> 행수, 열수, 해당순서(1행 2열의 차트 표시)
plt.plot(np.arange(5)**2)
plt.title('char1')
plt.subplot(1, 2, 2)
plt.plot(np.arange(5), np.arange(5)**3, 'bs--')
plt.title('chart2')
plt.tight_layout() #=> 두 차트 사이 간격 주기
plt.show()
````

````python
plt.subplot(2, 2, 1) #=> 행수, 열수, 해당순서(2행 2열의 차트 표시)
plt.plot(np.arange(5)**2)
plt.title('char1')

plt.subplot(2, 2, 2)
plt.plot(np.arange(5), np.arange(5)**3, 'bs--')
plt.title('chart2')

plt.subplot(2, 2, 3)
plt.plot(np.arange(5)**2, 'g')
plt.title('char3')

plt.subplot(2, 2, 4)
plt.plot(np.arange(5), np.arange(5)**3, '--')
plt.title('chart4')

plt.tight_layout() #=> 두 차트 사이 간격 주기
plt.show()
````

# 막대형 그래프
```python
plt.bar()
plt.barh()
```

````python
y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다']
plt.title('bar chart')

plt.xticks(x, xlabel) #=> plt.xlabel: x축 전체의 이름 / plt.xticks(x, xlabel): x축 각 값에 각각 라벨링
plt.yticks(sorted(y))
plt.bar(x,y) #=> plt.bar()에 x,y 값 적용
plt.show()

y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다']
plt.title('horizental bar chart')

plt.yticks(x, xlabel) #=> x y 축이 서로 바뀌었으므로 y축에 xlabel을 적용
plt.xticks(sorted(y))
plt.barh(x, y, alpha=0.3) #=> plt.barh()는 가로막대그래프, alpha는 투명도
plt.grid() #=> 배경에 그리드
plt.show()
````

# 파이차트
```python
plt.pie()
```

````python
labels = ['호박파이', '사과파이', '블루베리파이', '엄마손파이']

sizes = [15, 50, 72, 10]
colors = ['lightgreen', 'gold', 'lavender', 'lightcoral']

explode = (0, 0.1, 0.2, 0.3)

plt.title("파이 차트")

plt.pie(sizes, labels=labels, explode=explode, colors=colors,
         autopct='%1.1f%%', shadow=True, startangle=130)

# explode => 조각 사이 간격
# autopct => 비율 표시
# shadow => 그림자 유무
# startangle => 그래프가 반시계방향으로 돌아가는 각도

plt.axis('equal')

plt.show()
````

# pandas
- Pandas의 Series, DataFrame에 `plot` 시각화 메소드 내장
- `plot`은 matplotlib을 내부에 import하여 사용

````python
np.random.seed(2023)
df1 = pd.DataFrame(
    np.random.randn(100, 3), # 100 X 3 정규분포를 따르는 난수 배열
    index=pd.date_range('2023.12.12', periods=100), # 시작점, 갯수
    columns=['A', 'B', 'C'],
).cumsum() # 행들을 이전 행을 더한 값으로 누적합
# .cumprod() # 누적곱
df1
````

````python
df1.plot() #=> 데이터프레임에 내장된 plot 사용
plt.title('df 내장 플롯')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
````

````python
iris : pd.DataFrame = sns.load_dataset('iris') # 붓꽃
titanic : pd.DataFrame = sns.load_dataset('titanic') # 타이타닉호 데이터

# kind = 종류 => 해당 종류 plot 교체
iris['sepal_length'][:20].plot(
    kind='bar', rot=45, ylim=(4, 6), grid=True)
#=> kind=플롯종류, rot=축값 회전 각도, ylim=y축값제한((4,6)형식으로 작성), grid=그리드 유무, subplot=서브플롯 여부

iris[:20].plot(
    kind='bar', rot=45, grid=True, subplots=True)
plt.tight_layout() #=> 차트 사이 여유 간격 주기
````

````python
# kind 를 사용할 필요 없이 plot 다음에 직접적으로 해당 차트의 메소드 호출해서 사용
iris[:5].plot.bar(rot=0)
plt.ylim(0, 7)
````

````python
# 파이 차트 : 원형, 비율을 나타낼 때
df2 = titanic['class'].value_counts()
df2
df2.plot.pie(autopct="%.2f%%") # 포맷설정
plt.axis('equal')
plt.show()
````
